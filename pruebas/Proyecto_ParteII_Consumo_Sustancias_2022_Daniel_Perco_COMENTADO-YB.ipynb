"¿Cuáles son los principales factores predictivos del consumo problemático de sustancias en Argentina, y cómo se pueden utilizar estos factores para predecir las tendencias futuras y los posibles impactos en la salud pública y la sociedad?"

Esta pregunta se centra en identificar variables específicas que puedan servir como indicadores predictivos del consumo problemático de sustancias. Estas variables podrían incluir, por ejemplo, la frecuencia y cantidad de consumo, patrones de uso (como el uso recreativo versus el uso compulsivo), factores demográficos (edad, género, ubicación geográfica), factores socioeconómicos (nivel de ingresos, educación), y otros factores de riesgo (como historial de salud mental o acceso a sustancias). El objetivo es utilizar estas variables para predecir tendencias y riesgos futuros, permitiendo una intervención más temprana y efectiva

•	Determinar qué factores están más fuertemente asociados con el consumo de sustancias.

Impacto en Salud y Sociedad:
•	Evaluar cómo el consumo de sustancias afecta a diferentes aspectos de la salud y el bienestar social.

Desarrollo de Modelos Predictivos:

```python
# # Filtrar por sexo válido (1 o 2)
df_filtrado = df[df['SEXO_SEL'].isin([1, 2])]

# Agrupar por edad y sexo, y contar
conteo_por_edad_y_sexo = df_filtrado.groupby(['EDAD_SEL', 'SEXO_SEL']).size().unstack()

# Crear un gráfico de barras apiladas
ax = conteo_por_edad_y_sexo.plot(kind='barh', stacked=True, figsize=(10, 25))

# Agregar títulos y etiquetas
plt.title('Distribución de Género por Edad')
plt.xlabel('Cantidad')
plt.ylabel('Edad')
plt.legend(['Hombres', 'Mujeres'])

ax.set_facecolor("#BBB")

# Añadir etiquetas en cada barra
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{width}', (x + width/2, y + height/5), ha='center',color='white')
# Mostrar el gráfico
plt.show()
```

## Aplicación de Kmeans

Recordar que el valor de nuestra variable TARGET no debe **modificarse**

```python
# Esto es lo que deberíamos obtener si analizamos la variable "Target"
df.consumo_de_med_sin_receta.unique()
```

En este caso probamos con una estandarización de nuestros valores, sin demasiado pre-procesamiento. Básicamente tomando las columnas con datos numéricos y eliminando además la columna correspondiente al ID de la persona que no aporta valor en la clusterización en principio (_evaluar otras variables que no sean de significancia para el clusterizado_)

```python
# # Luego normalizamos los datos manteniendo nuestra variable target sin modificar
# scaler = StandardScaler()

# # Seleccionamos columnas numéricas
df_numericas = df.select_dtypes(include='number')

# # Variable objetivo o "target"
target = df_numericas['consumo_de_med_sin_receta']

# # Escalado sin modificar la variable objetivo
# df_scaled = scaler.fit_transform(df_numericas.drop('consumo_de_med_sin_receta', axis=1))

df = pd.DataFrame(df_numericas, columns=df.select_dtypes(include='number').columns.difference(['consumo_de_med_sin_receta']))

df['consumo_de_med_sin_receta'] = target

df.head()
```

```python
# Definir las features que voy a utilizar para analizar
X = df.drop(['consumo_de_med_sin_receta','ID_PER'], axis=1)
```

Instanciado de Modelo Kmeans

```python
# Instanciar el modelo
k_means = KMeans(n_clusters=5, n_init=10, random_state=42)

# Ajusto el modelo a mis datos
k_means.fit(X)

# Predecir
k_means_labels = k_means.predict(X)
```

### Visualizamos la clusterización obtenida

```python
# Visualizar los centroides
centroids = k_means.cluster_centers_
```

```python
df.describe().T
```

Dependiendo del PAR de variables X e Y seleccionadas, es como veremos la ubicación de los centroides con respecto a la Dispersión de Ptos (evaluar el par de variables que mejor se adapte para optimizar la visualización)

```python
# Gráfico de puntos mostrando los centroides
color = ['red', 'blue', 'green','orange','magenta']

# Creo nueva variable que contenga los valores obtenidos en el cluster
df['cluster'] = k_means_labels

# Creo Scatterplot 
sns.scatterplot(data=df, x='J_NIVEL_EDUCATIVO', y='J_EDAD', hue='cluster', palette=color)

# Graficar centroides
for i, centroid in enumerate(centroids):
  plt.scatter(centroid[0], centroid[1], c=color[i], s=200)

plt.title('Clusters de Consumo de Medicamentos sin Receta')
plt.xlabel('Nivel Educativo')
plt.ylabel('Edad')
plt.show()
```
