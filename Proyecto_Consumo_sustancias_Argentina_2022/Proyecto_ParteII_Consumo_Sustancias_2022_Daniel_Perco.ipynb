"¿Cuáles son los principales factores predictivos del consumo problemático de sustancias en Argentina, y cómo se pueden utilizar estos factores para predecir las tendencias futuras y los posibles impactos en la salud pública y la sociedad?"

Esta pregunta se centra en identificar variables específicas que puedan servir como indicadores predictivos del consumo problemático de sustancias. Estas variables podrían incluir, por ejemplo, la frecuencia y cantidad de consumo, patrones de uso (como el uso recreativo versus el uso compulsivo), factores demográficos (edad, género, ubicación geográfica), factores socioeconómicos (nivel de ingresos, educación), y otros factores de riesgo (como historial de salud mental o acceso a sustancias). El objetivo es utilizar estas variables para predecir tendencias y riesgos futuros, permitiendo una intervención más temprana y efectiva

Impacto en Salud y Sociedad:

```python
# # Filtrar por sexo válido (1 o 2)
df_filtrado = df[df['SEXO_SEL'].isin([1, 2])]

# Agrupar por edad y sexo, y contar
conteo_por_edad_y_sexo = df_filtrado.groupby(['EDAD_SEL', 'SEXO_SEL']).size().unstack()

# Crear un gráfico de barras apiladas
ax = conteo_por_edad_y_sexo.plot(kind='barh', stacked=True, figsize=(10, 25))

# Agregar títulos y etiquetas
plt.title('Distribución de Género por Edad')
plt.xlabel('Cantidad')
plt.ylabel('Edad')
plt.legend(['Hombres', 'Mujeres'])

ax.set_facecolor("#BBB")

# Añadir etiquetas en cada barra
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{width}', (x + width/2, y + height/5), ha='center',color='white')
# Mostrar el gráfico
plt.show()
```

# Analizmos  los dos consumos mayoritarios 

Veamos la variable Consumo de medicamentos sin Receta 

```python

from kmodes.kprototypes import KPrototypes


# Convertir el DataFrame a una matriz NumPy, que es necesaria para K-prototypes
X = df.values

# El índice de las columnas categóricas en tu matriz X
indices_categoricos = [df.columns.get_loc(c) for c in columnas_categoricas if c in df]

# Inicializar y ajustar K-prototypes
kproto = KPrototypes(n_clusters=3, init='Cao', verbose=2)
clusters = kproto.fit_predict(X, categorical=indices_categoricos)

# Añadir las etiquetas del cluster al DataFrame
df['cluster'] = clusters


```

```python
# Obtener centroides
centroides = kproto.cluster_centroids_

# Separar en centroides numéricos y categóricos
centroides_numericos = centroides[0]
centroides_categoricos = centroides[1]

# Imprimir centroides
for i in range(len(centroides_numericos)):
    print(f"Centroide {i}")
    print("Numérico:", centroides_numericos[i])
    
    # Asegurarse de que cada elemento en centroides_categoricos[i] es tratado individualmente
    print("Categórico:", centroides_categoricos[i])
    print()
```

```python
# visualizar los clusters

#'clusters' es  etiquetas de cluster y 'variable_categorica' es variable categórica de interés
df['cluster'] = clusters
matriz_confusion = pd.crosstab(df['cluster'], df['consumo_de_med_sin_receta'])

print(matriz_confusion)
```

```python

# 'X' es tu conjunto de datos y 'indices_categoricos' es la lista
# de índices de las columnas categóricas en 'X'
 

costos = []
K = range(1, 10)  # Por ejemplo, probarás de 1 a 9 clusters. ajustar este rango según sea necesario.
for k in K:
    kprototype = KPrototypes(n_clusters=k, init='Cao', n_jobs=-1, verbose=1)
    kprototype.fit_predict(X, categorical=indices_categoricos)
    costos.append(kprototype.cost_)

# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(K, costos, 'bx-')
plt.xlabel('k')
plt.ylabel('Costo')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.show()
```

```python
# GRAFICAMOS EL COEFICIENTE DE SILUETA PARA VARIOS CLUSTERS


from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Supongamos que 'X' es tu matriz de datos y 'range_n_clusters' es el rango de números de clusters que quieres probar.
range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:
    # Crear un subplot con 1 columna y 1 fila
    fig, ax1 = plt.subplots(1, 1)
    fig.set_size_inches(18, 7)

    # El coeficiente de silueta puede variar entre -1, 1 pero en este ejemplo lo limitaremos entre -0.1, 1
    ax1.set_xlim([-0.1, 1])
    # El (n_clusters+1)*10 es para insertar un espacio vacío entre la silueta
    # para destacar los clusters.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Inicializar el clusterer con el valor n_clusters y un random seed de 10 para la reproducibilidad.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # El coeficiente de silueta promedio para todos los puntos
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("Para n_clusters =", n_clusters,
          "El promedio del coeficiente de silueta es:", silhouette_avg)

    # Calcular las puntuaciones de silueta para cada muestra
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Agregar las puntuaciones de silueta para las muestras pertenecientes al cluster i, y ordenarlas
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Etiquetar los gráficos de silueta con el número del cluster en el medio
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Calcular el nuevo y_lower para el siguiente plot
        y_lower = y_upper + 10  # 10 para las 0 muestras

    ax1.set_title("El gráfico de silueta para los varios clusters.")
    ax1.set_xlabel("El coeficiente de silueta")
    ax1.set_ylabel("Etiqueta del cluster")

    # La línea vertical para el coeficiente de silueta promedio de todos los valores
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Limpiar las etiquetas del eje y
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    plt.suptitle(("Análisis de silueta para clustering KMeans sobre la muestra de datos "
                  "con n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

Interpretando tus gráficos de silueta:

k=2 Clusters: Observamos dos áreas que representan dos clusters, con siluetas bastante amplias, lo que sugiere una buena separación entre los clusters. Sin embargo, hay áreas de las siluetas que caen por debajo del promedio, lo que indica que algunos puntos podrían estar mal asignados o que los clusters no son muy distintos.

k=3 Clusters: Aquí vemos una variación más notable en las alturas de las siluetas, lo que indica diferencias en la cohesión dentro de los clusters. Un cluster (el del medio) parece tener una consistencia menor en comparación con los otros dos.

k=4 Clusters: La variabilidad en la altura de las siluetas aumenta, indicando posiblemente que estamos empezando a dividir en clusters que no son tan cohesivos. Sin embargo, no hay valores significativamente negativos que sugieran una asignación incorrecta.

k=5 y k=6 Clusters: Con más clusters, observamos que la consistencia de las siluetas disminuye, con alturas más variables dentro de cada cluster, lo que sugiere que algunos puntos pueden no encajar bien dentro de su cluster asignado.

Para datos categóricos, el coeficiente de silueta debe interpretarse con cuidado. A diferencia de los datos continuos, donde la distancia euclidiana es una medida intuitiva de disimilitud, las medidas de disimilitud en datos categóricos son menos directas y la interpretación de lo que significa una "buena" separación o cohesión puede ser más difícil de discernir.

Las conclusiones generales podrían ser:

Un menor número de clusters (como 2 o 3) parece proporcionar una estructura de cluster más robusta y clara.
A medida que aumenta el número de clusters, parece que los clusters se vuelven menos cohesivos y más dispersos, lo que puede indicar un sobreajuste.
